```python
import pandas as pd
from py_stringmatching import Levenshtein, Jaro
from py_stringmatching import Jaccard
```

Read csv file

```python
publications = pd.read_csv('publications.csv',  encoding='latin1')
```

```python
publications
```

Initialize the similarity measures

```python
levenshtein = Levenshtein()
jaro = Jaro()
jaccard = Jaccard()
```

Computing the pairwise similarity

```python
def compute_similarity(row1, row2):
    #Levenshtein similarity
    title_score = levenshtein.get_sim_score(row1['title'], row2['title'])
    #Jaro similarity
    authors_score = jaro.get_sim_score(row1['authors'], row2['authors'])
    return title_score, authors_score
```

Compare each pair of records

```python
for i, row_i in publications.iterrows():
    for j, row_j in publications.iterrows():
        if i < j:  # To avoid comparing a record with itself and comparing the same pairs twice
            # Convert to string to ensure compatibility with the similarity functions
            title_i, authors_i = str(row_i['title']), str(row_i['authors'])
            title_j, authors_j = str(row_j['title']), str(row_j['authors'])

            s_t = levenshtein.get_sim_score(title_i, title_j)
            s_a = jaro.get_sim_score(authors_i, authors_j)

            print(f'Comparing Record {i} and Record {j}: Title Similarity: {s_t}, Authors Similarity: {s_a}')
```

```python
# Define a threshold for considering records as duplicates
similarity_threshold = 0.7

# List to store the IDs of duplicate records
duplicate_records = []

for i, row_i in publications.iterrows():
    for j, row_j in publications.iterrows():
        if i < j:  # To avoid comparing a record with itself and comparing the same pairs twice
            # Convert to string to ensure compatibility with the similarity functions
            title_i, authors_i = str(row_i['title']), str(row_i['authors'])
            title_j, authors_j = str(row_j['title']), str(row_j['authors'])
            venue_i, venue_j = str(row_i['venue']).split(), str(row_j['venue']).split()

            s_t = levenshtein.get_sim_score(title_i, title_j)
            s_a = jaro.get_sim_score(authors_i, authors_j)

            # Use the modified affine similarity for the conference (venue) attribute (s c)
            s_c = jaccard.get_sim_score(venue_i, venue_j)

            # Use Match (1) / Mismatch (0) for the year (s y)
            s_y = 1 if row_i['year'] == row_j['year'] else 0

            # Calculate the final similarity score
            rec_sim = 0.5 * s_t + 0.2 * s_a + 0.2 * s_c + 0.1 * s_y

            # Check if the record pair is a duplicate
            if rec_sim > similarity_threshold:
                duplicate_records.append((row_i['id'], row_j['id']))


            #print(f'Comparing Record {i} and Record {j}: Title Similarity: {s_t}, Authors Similarity: {s_a}')

#Print the list of duplicate records
print(duplicate_records)
```

Load pub_mappings to count the number of duplicates records discovered correctly

```python
actual_mappings= pd.read_csv('pub_mappings.csv')
```

```python
# Convert the 'idDBLP' column to a set
idDBLP_set = set(actual_mappings['idDBLP'])

duplicate_ids = {id for record_pair in duplicate_records for id in record_pair}

# Count the number of correctly identified duplicate IDs
correctly_identified_ids = duplicate_ids.intersection(idDBLP_set)
num_correctly_identified = len(correctly_identified_ids)

# Output the results
print(f"Number of correctly identified duplicate IDs: {num_correctly_identified}")
print(f"Total number of unique IDs in duplicate records: {len(duplicate_ids)}")
```
