**Question 3**

 Function to print the column names with missing values


```python
def report_missing_values(df, table_name):
    missing_values = df.isna().any()
    columns_with_missing_values = missing_values[missing_values].index.tolist()
    print(f"The following columns in the '{table_name}' table have missing values:")
    for col in columns_with_missing_values:
        print(col)
    return columns_with_missing_values
```

```python
missing_columns = report_missing_values(merged_dataset, 'merged_dataset')
```

Replace the missing values in columns with -1

```python
merged_dataset[missing_columns] = merged_dataset[missing_columns].fillna(-1)
```

**Question 4**

Function to remove outliers from the dataset

```python
def remove_outliers(df):
    # Start with an empty list to hold the indices of outliers
    indices_to_remove = []
    # For each numerical column, calculate the IQR and determine the outliers
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        # Define bounds for what is considered an extreme value
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Find the indices of outliers in the column and append to the list
        column_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
        indices_to_remove.extend(column_outliers)
    # Remove duplicates from the list of indices to remove
    indices_to_remove = list(set(indices_to_remove))

    # Drop the outliers from the DataFrame
    df_cleaned = df.drop(index=indices_to_remove)

    return df_cleaned, indices_to_remove
```

Function call to remove outliers from my merged_dataset

```python
merged_dataset_cleaned, outliers_indices = remove_outliers(merged_dataset)
```
